## Make the first Quantum Circuit
To make a quantum circuit with Qiskit C++, include following header file. 

````c++
#include "circuit/quantumcircuit.hpp"
````

Then you can make a quantum circuit by Qiskit C++. This example will make a quantum circuit with 2 qubits and 2 classical bits.

````c++
// by using namespace, you can write as similar to Qiskit Python
using namespace Qiskit::circuit;

// new quantum circuit with 2 qubits and 2 classical bits
QuantumCircuit circ(2, 2);
circ.h(0);
circ.z(1);
circ.measure(0, 0);
circ.measure(1, 1);
````

You can print out the list of gates on the quantum circuit by `print` function.

````c++
circ.print();
````

## Using Registers
`QuantumRegister` and `ClassicalRegister` classes are used to name the registers. The qubits and classical bits can be grouped into some register classes and can be used to identify instead of using global bits numbers. It is usefull for grouping classical bits for measurements.

This example makes a quantum circuit from `QuantumRegister` and `ClassicalRegister` classes.

````c++
using namespace Qiskit::circuit;

// new quantum register with 2 qubits
auto qreg = QuantumRegister(2);
// new classical register with 2 bits named as measure
auto creg = ClassicalRegister(2, std::string("measure"));
// new classical register with 2 bits named as test
auto ctest = ClassicalRegister(2, std::string("test"));
// make a new quantum circuit with registers
QuantumCircuit circ_with_reg(std::vector<QuantumRegister>({qreg,}), std::vector<ClassicalRegister>({creg, ctest}));

circ_with_reg.measure(qreg, ctest);
circ_with_reg.h(qreg[0]);
circ_with_reg.x(qreg[1]);
circ_with_reg.measure(qreg, creg);

circ_with_reg.print();
````

Also you can print out the quantum circuit as QASM3 format.

````c++
std::cout << circ_with_reg.to_qasm3() << std::endl;
````
